```python
# Define the coordinates of Warrnambool
warrnambool_coords = (-38.3720954, 142.4777782)

# Function to calculate distance between each row and Warrnambool
def calculate_distance(row):
    location_coords = (row['Latitude'], row['Longitude'])
    return geodesic(warrnambool_coords, location_coords).miles

# Add a new "distance" column to the DataFrame
df['distance'] = df.apply(calculate_distance, axis=1)
df.sort_values(by='distance', ascending=True, inplace=True)

# Save the DataFrame back to a CSV file
df.to_csv('wb_dist.csv', index=False)
```

```python
df = pd.read_csv('/Users/gayathri/Documents/nj-local-news-analysis/aust_data/filtered_data.csv')
print(len(df))
df.head()
```

```python
# Calculate the map center and zoom level as before
Warmambool_center = [-38.3720954,142.4777782]
map_zoom = 5
mymap = folium.Map(location=Warmambool_center, zoom_start=map_zoom)

# Set a single color for all the markers
color = "#2F539B" 

def get_marker_size(count):
    scale_factor = 5  
    return 5 + (count / max_count) * scale_factor

# Find the maximum count to use for scaling the marker size
max_count = df["sum"].max()

# Add circle markers with varying sizes based on the count
for _, row in df.iterrows():
    location = row["GPE"]
    latitude = row["Latitude"]
    longitude = row["Longitude"]
    count = row["sum"]
    marker_size = get_marker_size(count)
    popup_text = f"Location: {location}\nCount: {count}"
    folium.CircleMarker([latitude, longitude], popup=popup_text, radius=marker_size, color=color, fill=True, fill_color=color).add_to(mymap)

# Saving the map
mymap.save("plot_warr.html")
```

```python
import folium

# Calculate the map center and zoom level as before
Warrnambool_center = [-38.3720954, 142.4777782]
map_zoom = 5
mymap = folium.Map(location=Warrnambool_center, zoom_start=map_zoom)

# Iterate through the DataFrame and add pinpoint markers
for _, row in df.iterrows():
    location = row["GPE"]
    latitude = row["Latitude"]
    longitude = row["Longitude"]
    popup_text = f"Location: {location}"
    
    # Create a pinpoint marker
    folium.Marker([latitude, longitude], tooltip=popup_text).add_to(mymap)

# Saving the map
mymap.save("pinpoint_visualization_aust.html")
```
